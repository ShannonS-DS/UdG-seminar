```python
from twython import Twython, TwythonStreamer
import json
```

```python
# Register app to https://apps.twitter.com/ , get CONSUMER_KEY, CONSUMER_SECRET, ACCESS_TOKEN, ACCESS_SECRET
# and save them in the conf/app.json file

with open('./conf/app.json', 'r') as app_file:
    app = json.load(app_file)
```

```python
# Twitter REST API
twitter = Twython(app["CONSUMER_KEY"], app["CONSUMER_SECRET"], oauth_version=2)
ACCESS_TOKEN = twitter.obtain_access_token()

twitter = Twython(app["CONSUMER_KEY"], access_token=ACCESS_TOKEN)

# Girona lat, lng, 2km
lat = 41.98311
lng = 2.82493
r = 2
geocode = str(lat) +","+ str(lng) +","+ str(r)+ "km"

# Other queries 
geo_located_tweets = twitter.search(geocode=geocode, count=100, result_type="recent")
```

```python
tweet = geo_located_tweets[u'statuses'][0]
print(json.dumps(tweet, indent=2))
```

```python
for tweet in geo_located_tweets[u'statuses']:
    print(tweet["user"]["name"] +" "+ tweet[u'created_at'])
    print(tweet[u'text']+"\n")
```

```python
# Twitter Streaming API 

class MyStreamer(TwythonStreamer):
    def on_success(self, tweet):
        print(tweet["user"]["name"] +" "+ tweet[u'created_at'])
        print(tweet[u'text']+"\n")
        with open('./data/tweets.json','a') as json_file:
            json.dump(tweet,json_file)
            json_file.write('\n')

            
boundingbox = "0.1591811,40.5230466,3.3325539,42.8614502"
stream = MyStreamer(app["CONSUMER_KEY"], app["CONSUMER_SECRET"],app["ACCESS_TOKEN"], app["ACCESS_SECRET"])
stream.statuses.filter(locations=boundingbox)
```
